```python
#Loading the Required Library
import numpy as np
import pandas as pd
from sklearn import preprocessing
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings("ignore")  #--to ignore warnings
```

```python
# loading dataset
dataset = pd.read_csv('C:/Users/Akaash/Downloads/cancer.csv')
dataset.head()
```

```python
#Complete Cancer dataset - applying lable encoder to species column - (Y Variable is Categorial)
label_encoder = preprocessing.LabelEncoder()
dataset['diagnosis'] = label_encoder.fit_transform(dataset['diagnosis']) 
```

```python
# split data into X and y
X = dataset.iloc[:,1:]
Y = dataset.iloc[:,0]
```

```python
# split data into train and test sets
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

#### Grid Search CV

```python
#importing The Required library
from sklearn.model_selection import GridSearchCV

#Creating SVM instance
clf = XGBClassifier()

#Defining Grid Search Parameter
param_grid = [{'n_estimators':[70,80,90,100],'max_depth':[3,4,5],'learning_rate':[0.1,0.2] }]

#Creating Grid Search instance and fitting the Model
gsv = GridSearchCV(clf,param_grid,cv=5)
gsv.fit(X_train,y_train)
```

```python
#Getting Search Results From Grid Search
gsv.best_params_ , gsv.best_score_ 
```

```python
# fit model no training data
model = XGBClassifier(n_estimators = 90, max_depth = 3,learning_rate=0.2)
model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
predictions
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

Inference: Using XGBM with Parameter n_estimators = 90 ans max_dept = 3,learning_rate=0.2 Comes the Best Accuracy of 97.34%

#### Light GBM ~ Light Gradient Boosting Machine

#### Installing the LGBM library

```python
#loading the Required library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# loading dataset
dataset = pd.read_csv('C:/Users/Akaash/Downloads/cancer.csv')
dataset.head()
```

```python
#Complete Cancer dataset - applying lable encoder to species column - (Y Variable is Categorial)
label_encoder = preprocessing.LabelEncoder()
dataset['diagnosis'] = label_encoder.fit_transform(dataset['diagnosis']) 
```

```python
# split data into X and y
X = dataset.iloc[:,1:]
Y = dataset.iloc[:,0]
```

```python
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# Importing the required Library
import lightgbm as lgb
from lightgbm import LGBMClassifier
#Creating the LGBM Dataset (Gradient Dataset)
d_train = lgb.Dataset(x_train, label=y_train)
```

#### Grid Search CV

```python
#importing The Required library
from sklearn.model_selection import GridSearchCV

#Creating SVM instance
clf = lgb.LGBMClassifier(boosting_type = 'gbdt', objective = 'binary')
#Defining Grid Search Parameter
param_grid = [{'n_estimators':[70,80,100],'num_leaves':[5,10,15],'max_depth':[4,5,6],'learning_rate':[0.1,0.2] }]

#Creating Grid Search instance and fitting the Model
gsv = GridSearchCV(clf,param_grid,cv=5)
gsv.fit(x_train,y_train)
```

```python
#Getting Search Results From Grid Search
gsv.best_params_ , gsv.best_score_ 
```

```python
#Defining the Parameters for LGBM
params = {}
params['learning_rate'] = 0.2
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 15
params['min_data'] = 50
params['max_depth'] = 4
```

```python
#Training the Lgbm Dataset 
clf = lgb.train(params, d_train, 70) # n_estimators = 70
```

```python
#Prediction
y_pred=clf.predict(x_test)
```

```python
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
accuracy*100
```

Inference: Using LGBM with Parameter n_estimators = 70 ans max_dept = 4, Comes the Best Accuracy of 97.90%
